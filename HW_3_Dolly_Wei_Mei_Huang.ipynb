```
# df['graffiti_flag']= df.apply(recode_graffitti_count, axis=1)
def recode_graffiti_counts(row):
  if pd.notnull(row.descriptor):
    if 'graffiti' in row.complaint_type.lower():
      return True 
    elif 'graffiti' in row.descriptor.lower():
      return True
    else:
      return False
  
```

```
type(recode_graffiti_counts)
```

```
# Step 2
df['graffiti_flag'] = df.apply(recode_graffiti_counts, axis=1)
```
